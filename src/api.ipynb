```python
type_failure = (1,10)
time_repair = (0.1,2)
cost = (0,2)
criticality = (0,0.9999)
humid = (0,100)
temp = (0,500)
```

```python
def generate_new_param():
  params = {
    "type_of_failure": random.uniform(type_failure[0], type_failure[1]),
    "time_repair": random.uniform(time_repair[0], time_repair[1]),
    "cost": random.uniform(cost[0], cost[1]),
    "criticality": random.uniform(criticality[0], criticality[1]),
    "humid": random.uniform(humid[0], humid[1]),
    "temp": random.uniform(temp[0], temp[1])
  }
  return params
```

```python
# print(f'Status \t Result')
for j in range(0,3):
    new_results = []
    for i in range(0,500):
        new_data = generate_new_param()
        response = requests.post(url+'/predict',data=json.dumps(new_data))
        result = response.json()
        predict = 'OK' if result['prediction'] == 0 else 'FAIL'
        # print(f'{predict}')
        print(f'{predict} \t {(result["score"]*100):.2f}%')
        new_data['label'] = (result['prediction'])
        # print(new_data)
        new_results.append(new_data)
        # data2.append(new_data, ignore_index=True)
        # pd.concat(data2)
        
    new_df = pd.DataFrame(data=new_results, columns=new_results[0].keys())

    data3 = pd.concat([new_df, data], ignore_index=True)

    data3.to_csv('../../ml_datasets/dataset_testing.csv', index=False)

    print(f"New data interation {j}")
    print("Update api")
    time.sleep(120)
    data2 = data3.sample(9000)
```
